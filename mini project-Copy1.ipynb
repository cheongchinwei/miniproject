```python
import pandas as pd
path = '04-23-2020.csv'
covid = pd.read_csv(path)
covid.isnull().sum() ##check missing value
```

```python
import pandas as pd
path2 = '04-14-2020.csv'
covid2 = pd.read_csv(path2)
covid2.isnull().sum() ##check missing value
```

```python
## Top 10 countries data (Last Update, Country/Region, Confirmed, Deaths, Recovered) of Novel Coronavirus (COVID-19)
## date 14-Apr
covid2 = covid2.groupby(['Country_Region'])['Confirmed', 'Deaths', 'Recovered'].sum().sort_values("Confirmed", ascending=False)
pass_covid = covid2.head(10)
print (pass_covid)
```

```python
## Top 10 countries data (Last Update, Country/Region, Confirmed, Deaths, Recovered) of Novel Coronavirus (COVID-19)
## date 23-Apr
covid = covid.groupby(['Country_Region'])['Confirmed', 'Deaths', 'Recovered'].sum().sort_values("Confirmed", ascending=False)
latest_covid = covid.head(10)
print (latest_covid)
```

```python
## Date: 23-Apr
import matplotlib.pyplot as plt

bar_color = ['yellow'] * len(latest_covid)
bar_color [0] = 'b'
bar_color [6] = 'r'
bar_color [8:10] = ['red']*2
covid3 = latest_covid.groupby(by='Country_Region')['Confirmed'].sum().sort_values(ascending=False).plot(kind='bar', color = bar_color)
fig = plt.gcf()
fig.set_size_inches(10,5)

covid3.set_xlabel ('Top 10 Country')
covid3.set_ylabel ('Confirmed Cases')
covid3.set_title ('23-Apr-20 Covid-19 Data')
```

```python
## Date: 14-Apr
import matplotlib.pyplot as plt

bar_color = ['yellow'] * len(pass_covid)
bar_color [0] = 'b'
bar_color [6] = 'r'
bar_color [8:10] = ['red']*2
covid1 = pass_covid.groupby(by='Country_Region')['Confirmed'].sum().sort_values(ascending=False).plot(kind='bar', color = bar_color)
fig = plt.gcf()
fig.set_size_inches(10,5)

covid1.set_xlabel ('Top 10 Country')
covid1.set_ylabel ('Confirmed Cases')
covid1.set_title ('14-Apr-20 Covid-19 Data')
```

```python
import pandas as pd
path = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-11-2020.csv'
covid19 = pd.read_csv(path)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
import datetime

daily_trend_covid19 = pd.DataFrame(columns=['Date', 'Confirmed'])

start_date = datetime.date(2020, 3, 10)
end_date = datetime.date(2020, 4, 24)
delta = datetime.timedelta(days=1)

while start_date <= end_date:
    date_str = start_date.strftime('%m-%d-%Y')
    start_date += delta
    
    path = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/'+ date_str +'.csv'
    daily_covid19 = pd.read_csv(path)
    
    confirmed_sum = daily_covid19['Confirmed'].sum()
    
    daily_trend_covid19 = daily_trend_covid19.append({'Date': date_str, 'Confirmed' : confirmed_sum}, ignore_index=True)

fig, ax = plt.subplots()
ax.plot(daily_trend_covid19['Date'], daily_trend_covid19['Confirmed'])

plt.xticks(rotation=90)
fig.set_size_inches(18, 6)
ax.set_xlabel('Date')
ax.set_ylabel('Confirmed')
ax.set_title('Worldwide Confirmed Novel Coronavirus (COVID-19) cases over time')
```
